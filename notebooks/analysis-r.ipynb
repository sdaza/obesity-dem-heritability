```r
library(data.table)
library(relaimpo)
options(scipen = 100)
```

```r
dat = fread("../output/data/results-efast.csv")
dat = dat[iteration <= 3000]
dat = dat[population > 1000]
head(dat)
```

```r
dat[, i3 := random_mating * leakage * fertility_factor]
dat[, leakage_random_mating := leakage * random_mating]
dat[, leakage_fertility := leakage * fertility_factor]
dat[, random_mating_fertility := fertility_factor* random_mating]

m1 = lm(g4 ~ random_mating + leakage + fertility_factor + 
    random_mating:leakage + fertility_factor:leakage + 
    random_mating:fertility_factor, data = dat)

```

```r
summary(m1)
calc.relimp(m1)
```

```r
bootresult = boot.relimp(m1, b = 1000, 
    type = c("lmg", "pmvd", "pratt", "last", "first"), fixed = FALSE, rela = TRUE, 
        a)
```

```r
booteval.relimp(bootresult, typesel = c("lmg", "pmvd", "pratt", "last", "first"), 
    level = 0.9, bty = "perc")
```
